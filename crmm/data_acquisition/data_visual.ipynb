```python
import pandas as pd
dataset_name = 'cr2'
train = pd.read_csv(f'../../data/{dataset_name}_sec_6/train.csv')
test = pd.read_csv(f'../../data/{dataset_name}_sec_6/test.csv')
val = pd.read_csv(f'../../data/{dataset_name}_sec_6/val.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_sample_distribution(df, title,  column_name='Rating', figsize=(10, 6), label_mapping=None):
    """
    绘制数据集的样本分布图。

    参数:
    df (pd.DataFrame): 包含样本标签的数据集
    column_name (str): 样本标签所在列的名称，默认为'Rating'
    figsize (tuple): 图形大小，默认为(10, 6)
    label_mapping (dict): 将数值下标映射到文字标签的字典，默认为None
    """
    counts = df[column_name].value_counts()

    plt.figure(figsize=figsize)
    bars = plt.bar(counts.index, counts.values)
    plt.xlabel(column_name)
    plt.ylabel('frequency')
    plt.title(f'sample distribution of {title}')

    if label_mapping:
        plt.xticks(list(label_mapping.keys()), list(label_mapping.values()))

    # 在柱顶部添加数量标签
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, str(height), ha='center', va='bottom')

    plt.show()

```

```python
label_mapping = {
    0: 'AA+',
    1: 'A',
    2: 'BBB',
    3: 'BB',
    4: 'B',
    5: 'CCC-'
}
plot_sample_distribution(train, 'train', label_mapping=label_mapping, figsize=(6, 4))
plot_sample_distribution(test, 'test', label_mapping=label_mapping, figsize=(6, 4))
plot_sample_distribution(val, 'val', label_mapping=label_mapping, figsize=(6, 4))
```
