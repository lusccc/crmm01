```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../../data/cr_sec_ori/corporate_rating.csv')
data
```

```python
symbols = data['Symbol']
```

```python
from cik import getCIK
ciks = []
for sb in symbols:
    ciks.append(str(getCIK(sb)))
```

```python
data['CIK'] = ciks
```

```python
data['Id'] = data.index
data = data.set_index('Id')
data
```
